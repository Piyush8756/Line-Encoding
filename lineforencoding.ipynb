```python
#!/usr/bin/env python
# coding: utf-8

# In[13]:

#Piyush verma 
#Rakesh Kumar
import numpy as np
import matplotlib.pyplot as plt

print("choose any number between 1 to 5 of given below which you want to implement")
print("\n1.Implementaion of NRZ-L\n")
print("2.Implementaion of NRZ-I\n")
print("3.Implementaion of Manchester\n")
print("4.Implementaion of Differantial Manchester\n")
print("5.Implementaion of anyone of these(AMI/B8ZS/HDB3)\n")
n=int(input())
print("Enter the binary number to plot graph")
s=input()
st = s

if(n==1):
    print(" NRZ-L ")
    ls=list()
    for i in range(len(s)):
        if(s[i]=='0' or s[i]==0):
            ls.append(-1)
        else:
            ls.append(1)
    xs = np.repeat(range(len(s)), 2)
    ys = np.repeat(ls, 2)
    xs=xs[1:]
    xs=np.append(xs,(xs[len(xs)-1]+1))
    ys=ys[:-1]
    ys=np.append(ys,(ys[len(ys)-1]))
    plt.step(xs,ys)
    plt.ylim(-2,2)
    plt.show()
    
    
elif(n==2):
    print("NRZ-I")
    p=list()
    if(s[0]=='0' or s[0]==0):
        p.append(-1)
    else:
        p.append(1)
    l=len(s)
    i=1
    while(i<l):
        if(int(s[i])==0):
            p.append(p[i-1])
        else:
            p.append(-p[i-1])
        i=i+1
    xs = np.repeat(range(len(s)), 2)
    ys = np.repeat(p, 2)
    xs=xs[1:]
    xs=np.append(xs,(xs[len(xs)-1]+1))
    ys=ys[:-1]
    ys=np.append(ys,(ys[len(ys)-1]))
    plt.step(xs,ys)
    plt.ylim(-2,2)
    plt.show()
   
elif(n==3):
    print("Manchester")
    q=list()
    for j in range(len(s)):
        if(s[j]=='0' or s[j]==0):
            q.append(-1)
            q.append(1)
        else:
            q.append(1)
            q.append(-1)
    xs=[x*0.5 for x in range(0,(2*len(s)))]
    xs=np.repeat(xs,2)
    ys = np.repeat(q, 2)
    xs=xs[1:]
    xs=np.append(xs,(xs[len(xs)-1]+0.5))
    ys=ys[:-1]
    ys=np.append(ys,(ys[len(ys)-1]))
    plt.step(xs,ys)
    plt.ylim(-2,2)
    plt.show()

elif(n==4):
    print("Differantial Manchester")
    r=list()
    r.append(1)
    r.append(-1)
    i=1
    k=len(s)
    while(i<k):
        if(int(s[i])==1):
            r.append(r[len(r)-1])
            r.append(-r[len(r)-1])
        else:
            r.append(-r[len(r)-1])
            r.append(-r[len(r)-1])
        i=i+1
    print(r)
    xs=[x*0.5 for x in range(0,(2*len(s)))]
    xs=np.repeat(xs,2)
    ys = np.repeat(r, 2)
    xs=xs[1:]
    xs=np.append(xs,(xs[len(xs)-1]+0.5))
    ys=ys[:-1]
    ys=np.append(ys,(ys[len(ys)-1]))
    plt.step(xs,ys)
    plt.ylim(-2,2)
    plt.show()
    
else:
    print("0.AMI","1.HDB3 or B8ZS")
    q=int(input())
    if(q==0):
        print("Perform AMI")
        am=list()
        m=1
        for i in range(len(s)):
            if(int(s[i])==0):
                am.append(0)
            else:
                if(m%2==1):
                    am.append(1)
                else:
                    am.append(-1)
                m=m+1
        xs = np.repeat(range(len(s)), 2)
        ys = np.repeat(am, 2)
        xs=xs[1:]
        xs=np.append(xs,(xs[len(xs)-1]+1))
        ys=ys[:-1]
        ys=np.append(ys,(ys[len(ys)-1]))
        plt.step(xs,ys)
        plt.ylim(-2,2)
        plt.show()
       
    else:
        print("implementaion of Scrambling")
        print("1.B8ZS","2.HDB3")
        p=int(input())
        q=len(s)
        if(p==1):
            print("B8ZS")
            b=list()
            m=1
            s1=s.replace("00000000","000vb0vb")
            for i in range(len(s1)):
                if(s1[i]=='0' or s1[i]==0):
                    b.append(0)
                elif(s1[i]=='1'):
                    if(m%2==1):
                        b.append(1)
                    else:
                        b.append(-1)
                    m=m+1
                elif(s1[i]=='v'):
                    if(m%2==1):
                        b.append(-1)
                    else:
                        b.append(1)
                else:
                    if(m%2==1):
                        b.append(1)
                    else:
                        b.append(-1)
                    m=m+1
            xs = np.repeat(range(len(s)), 2)
            ys = np.repeat(b, 2)
            xs=xs[1:]
            xs=np.append(xs,(xs[len(xs)-1]+1))
            ys=ys[:-1]
            ys=np.append(ys,(ys[len(ys)-1]))
            plt.step(xs,ys)
            plt.ylim(-2,2)
            plt.show()
              
        else:
            print("HDB3")
            m=0
            h=list()
			  
            f=s.find("0000")
            if(f==-1):
                f=len(s)
            i=0
            k=len(s)
            d=1
            p=0
            while(i<k):
                if(s[i]=='1' or s[i]==1):
                    m=m+1
                    p=p+1
                    if(m%2==1):
                        h.append(d)
                        d=1
                    else:
                        h.append(-d)
                        d=-d
                else:
                    if(i<f):
                        h.append(0)
                    elif(i==f):
                        i=i+3
                        if(p%2==0):
                            h.append(-d)
                            h.append(0)
                            h.append(0)
                            h.append(-d)
                            d=-d
                            p=p+2
                            m=m+1
                        else:
                            h.append(0)
                            h.append(0)
                            h.append(0)
                            h.append(d)
                            p=p+1
                        jk=s[i+1:(i+1)+(k-i-1)]
                        x=jk.find("0000")
                        if(x==-1):
                            f=k
                        else:
                            f=i+1+x
                i=i+1
            xs = np.repeat(range(len(s)), 2)
            ys = np.repeat(h, 2)
            xs=xs[1:]
            xs=np.append(xs,(xs[len(xs)-1]+1))
            ys=ys[:-1]
            ys=np.append(ys,(ys[len(ys)-1]))
            plt.step(xs,ys)
            plt.ylim(-1.5,1.5)
            plt.show()
            print(h)
```
